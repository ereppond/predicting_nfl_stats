```python
def rename_columns(df_orig):
    '''returns df with renamed columns'''
    return df_orig.rename(index=str, columns={'YearMade': 'year_made', 'saledate': 'sale_date', 'MachineHoursCurrentMeter': 'machine_hours', 
                         'UsageBand': 'usage_band', 'ProductSize': 'product_size', 'Engine_Horsepower': 'horsepower', 
                         'Blade_Width': 'blade_width', 'Hydraulics': 'hydraulics', 'Tire_Size': 'tire_size',
                         'Undercarriage_Pad_Width': 'pad_width', 'Stick_Length': 'stick_length','Thumb':'thumb',
                         'Pattern_Changer': 'pattern_changer', 'Grouser_Type': 'grouser_type', 'Backhoe_Mounting': 'backhoe_mounting',
                         'Blade_Type': 'blade_type', 'Travel_Controls': 'travel_controls', 'Differential_Type': 'differential_type',
                         'Steering_Controls': 'steering_controls', 'SalesID': 'sales_id', 'SalesPrice': 'sales_price', 'MachineID': 'machine_id',
                         'ModelID': 'model_id', 'datasource': 'data_source', 'auctioneerID': 'auctioneer_id', 'SalePrice': 'sale_price'})
```

```python
df = rename_columns(df_orig)
```

```python
def nan_to_none(df):
    df['hydraulics'].fillna('None or Unspecified', inplace=True)
    df['product_size'].fillna('None', inplace=True)
    df['blade_width'].fillna('None', inplace=True)
    df['stick_length'].fillna('None', inplace=True)
    df['pad_width'].fillna('None', inplace=True)
    df['thumb'].fillna('None', inplace=True)
    df['stick_length'].fillna('None', inplace=True)
    df['pattern_changer'].fillna('None', inplace=True)
    df['grouser_type'].fillna('None', inplace=True)
    df['backhoe_mounting'].fillna('None', inplace=True)
    df['blade_type'].fillna('None', inplace=True)
    df['travel_controls'].fillna('None', inplace=True)
    df['differential_type'].fillna('None', inplace=True)
    df['steering_controls'].fillna('None', inplace=True)
```

```python
nan_to_none(df)
```

```python
def nan_to_zero(df):
    for col, value in df.iteritems():
        df[col].fillna(0, inplace=True)
```

```python
nan_to_zero(df)
```

```python
#turns sale_date into datetime object
df['true_date'] = pd.to_datetime(df['sale_date'])
```

```python
df = df[df['year_made'] > 1800]
```

```python
df['age_of_product'] = df['true_date'].apply(lambda x: x.year) - df['year_made']
```

```python
df_num = df.loc[:,['year_made', 'machine_hours', 'age_of_product']]
```
